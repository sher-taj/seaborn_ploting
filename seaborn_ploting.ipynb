### Box plot

```python
df['class'].unique()
```

```python
# set the theme

sns.set_theme(style='darkgrid',font_scale=1,font='Times New Roman')

plt.figure(figsize=(12,6)) # A 4 Size 16:9
sns.boxplot(data=df, x='sex',y='age',hue='class',palette={'First':'red','Second':'green','Third':'blue'})

# Label

plt.xlabel('age')
plt.ylabel('fare')
plt.title("Class and Age Base Fare")
# change the legend position
plt.legend(loc='upper left',title='Class Name')
```

```python
# set the theme

sns.set_theme(style='whitegrid',font_scale=1,font='Times New Roman')

plt.figure(figsize=(12,6)) # A 4 Size 16:9
sns.boxenplot(data=df, x='sex', y='age', hue='class',  palette={'First': 'red', 'Second': 'green', 'Third': 'blue'}, order=['male', 'female'])

# Label
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title("Boxen Plot of Age Based on Gender and Class Name of Passenger")

# Change the legend position
plt.legend(loc='upper left', title='Class Name')

# Change the X-Ticks Name
plt.xticks(ticks=[0, 1], labels=["Male", "Female"], rotation=45)

plt.show()
```

# Bar Plot

```python
# set the theme

sns.set_theme(style='whitegrid',font_scale=1,font='Times New Roman')

plt.figure(figsize=(12,6)) # A 4 Size 16:9
sns.barplot(data=df, x='sex', y='age', hue='class',  palette={'First': 'red', 'Second': 'green', 'Third': 'blue'}, order=['male', 'female'],capsize=0.5)

# Label
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title("Bar Plot of Age Based on Gender and Class Name of Passenger")

# Change the legend position
plt.legend(loc='upper left', title='Class Name')

# Change the X-Ticks Name
plt.xticks(ticks=[0, 1], labels=["Male", "Female"], rotation=45)

```

```python
# set the theme

sns.set_theme(style='whitegrid',font_scale=1,font='Times New Roman')

plt.figure(figsize=(12,6)) # A 4 Size 16:9
sns.violinplot(data=df, x='sex', y='age', hue='class',  palette={'First': 'red', 'Second': 'green', 'Third': 'blue'}, order=['male', 'female'])

# Label
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title("Bar Plot of Age Based on Gender and Class Name of Passenger")

# Change the legend position
plt.legend(loc='upper left', title='Class Name')

# Change the X-Ticks Name
plt.xticks(ticks=[0, 1], labels=["Male", "Female"], rotation=45)

```
